```python
import os
import pandas as pd

# Load in the dataset 
currentfolder = os.getcwd()

# Load in the dataset
ts = pd.read_csv(currentfolder + '/other_data/saga_merged.csv', index_col=0)	
ts
```

```python
# Plot the timeseries of single ID, one kinematic feature and envolope

import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Make a subset of the data for participant V08

ts_s = ts[ts['ID'] == 'V08']

# Create subplots with dual y-axes
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(go.Scatter(x=ts_s['time'], y=ts_s['ENV'], name='ENV', mode='lines'))
fig.add_trace(go.Scatter(x=ts_s['time'], y=ts_s['acceleration_LEFT_INDEX'], name='acceleration_LEFT_INDEX', mode='lines'), secondary_y=True)

# Update layout
fig.update_layout(title='Multimodal timeseries', xaxis=dict(title='time (ms)'))

# Update y-axes titles and range
fig.update_yaxes(title_text='ENV', row=1, col=1, range=[-max(ts['ENV']), max(ts['ENV'])])
fig.update_yaxes(title_text='acceleration_LEFT_INDEX', row=1, col=1, secondary_y=True, range=[-max(ts['acceleration_LEFT_INDEX']), max(ts['acceleration_LEFT_ELBOW'])])

# Show xlim from 5000 to 10000
fig.update_xaxes(range=[5000, 10000])

# Get rid of the vertical and horizontal gridlines
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)

# Show the plot
fig.show()

```
